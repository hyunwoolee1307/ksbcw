```python
import xarray as xr
import xeofs
```

```python
da = xr.open_dataarray('data/line_208_interp_lf.nc')
```

```python
eof = xeofs.single.EOF(n_modes=10)
eof.fit(da, dim='date')
```

```python
comps = eof.components()
```

```python
scores = eof.scores()
```

```python
import matplotlib.pyplot as plt

plt.plot(da.date.values, scores[0])
```

```python
plt.plot(da.date.values, scores[1])
```

```python
plt.plot(da.date.values, scores[2])
```
